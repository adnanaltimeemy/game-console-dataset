```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Clean column names (removes spaces and standardizes format)
df.columns = df.columns.str.strip().str.lower()

# -------------------------------
# 4. Sales trends over time
# -------------------------------

if "sale_date" in df.columns:
    df["sale_date"] = pd.to_datetime(df["sale_date"], errors="coerce")

    sales_by_date = df.groupby("sale_date")["units_sold"].sum()

    plt.figure(figsize=(10,4))
    sales_by_date.plot(marker="o")
    plt.title("Units sold over time")
    plt.xlabel("Date")
    plt.ylabel("Units sold")
    plt.show()
else:
    print("Column 'sale_date' not found in dataset")

# -------------------------------
# 4a. Profit margin distribution
# -------------------------------

if "profit_margin" in df.columns:
    threshold = df["profit_margin"].median()   # example threshold

    plt.figure(figsize=(8,4))
    sns.histplot(df["profit_margin"], bins=20, kde=True)

    plt.axvline(threshold, color="red", linestyle="--",
                label=f"Threshold {threshold:.2f}")

    plt.title("Profit margin distribution")
    plt.xlabel("Profit margin")
    plt.legend()
    plt.show()
else:
    print("Column 'profit_margin' not found")

# -------------------------------
# 4b. Age distribution by profit level
# -------------------------------

if "profit_level" in df.columns and "customer_age" in df.columns:
    plt.figure(figsize=(8,4))
    sns.boxplot(x="profit_level", y="customer_age", data=df)

    plt.title("Customer age by profit level")
    plt.show()
else:
    print("Columns 'profit_level' or 'customer_age' not found")

# -------------------------------
# 4c. Gender vs profit level
# -------------------------------

if "gender" in df.columns and "profit_level" in df.columns:
    plt.figure(figsize=(6,4))
    sns.countplot(x="gender", hue="profit_level", data=df)

    plt.title("Gender vs profit level")
    plt.show()
else:
    print("Columns 'gender' or 'profit_level' not found")
```
